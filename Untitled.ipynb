```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pathlib

os.chdir('C:/Users/admin/PycharmProjects/Testing/Test1/')

HEIGHT = 350
def resize(img, height=HEIGHT, allways=False):
    if (img.shape[0] > height or allways):
        rat = height / img.shape[0]
        return cv2.resize(img, (int(rat * img.shape[1]), height))
    return img

#def auto_threshold(image, sigma=0.33):
#    v = np.median(image)
#    lower = int(max(0,(1.0 - sigma) * v))
#    upper = int(min(255,(1.0 + sigma) * v))
#    return lower, upper

def histthreshold(biImg):
    hist,bins = np.histogram(biImg.flatten(),256,[0,256])
    allpeaks = find_peaks(hist)
    #print(allpeaks)
    med = np.median(allpeaks)
    bol = np.array(allpeaks > med)
    filtered_list = [i for indx, i in enumerate(allpeaks) if bol[indx] == True]
    #plt.hist(biImg.ravel(),256,[0,256]); plt.show()
    #nwLen = len(filtered_list)
    #print(filtered_list)        
    lw, up = np.min(filtered_list), np.max(filtered_list)
    return lw, up


def find_peaks(a):
    x = np.array(a)
    max = np.max(x)
    lenght = len(a)
    ret = []
    for i in range(lenght):
        ispeak = True
        if i-1 > 0:
            ispeak &= (x[i] > 1 * x[i-1])
        if i+1 < lenght:
             ispeak &= (x[i] > 1 * x[i+1])

        ispeak &= (x[i] > 0.01 * max)
        if ispeak:
            ret.append(i)
    return ret

def mycont2(thr):
    global contours
    edges = cv2.Canny(imobj, thr, thr * 2)
    drawing = np.zeros(copyImg.shape, np.uint8)
    image, contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    objcorn = []
    for cnt in contours:
        rect = cv2.minAreaRect(cnt)
        points = cv2.boxPoints(rect)
        points = np.int0(np.around(points))
        objcorn.append(points)
#        print(objcorn)
#        cv2.polylines(drawing, [points], True, (255, 0, 0), 2)
        cv2.rectangle(drawing, tuple(points[0]), tuple(points[2]), (255,0,0), thickness=cv2.FILLED)
        cv2.imshow("output", drawing)
        cv2.imshow("rawImg", rawImg)
        cv2.imshow("input", imobj)
        cv2.imwrite('cheque_thr.png', drawing)

def mycont(thr):
    global contours
    edges = cv2.Canny(imobj, thr, thr * 2)
    drawing = np.zeros(copyImg.shape, np.uint8)
    image, contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        cv2.drawContours(drawing, [cnt], 0, (0, 255, 0), 2)
        cv2.imshow("output", drawing)
        cv2.imshow("rawImg", rawImg)
        cv2.imshow("input", imobj)
        cv2.imwrite('cheque_thr.png', drawing)
```
